            Required Analytical Questions
1.  What are the most popular colors by season?

```python
# Most popular color by season
color_counts = shopping_data.groupby(['Season', 'Color']).size().reset_index(name='count')
color_counts = color_counts.sort_values(['Season', 'count'], ascending=[True, False])
top_colors = color_counts.groupby('Season').head(1)
print(top_colors)
```

: In fall the most popular color is brown with a frequency of 117. 

: In Spring the top color is baby blue with a frequency of 83.

: For Summer lavender with a frequency of 120 and in winter black with a frequency of 155.

: Black was the most popular winter color with a frequency of 155.

: It may be benifical for the company to look into promotions and advertising that corresponds to the most popular color purchases for that particular season. For example promotions for items available in brown during the fall as well as advertising in store on the company website and social media.

2. What is the most popular clothing item by season?

```python
# Most popular clothing item by season
item_counts = shopping_data.groupby(['Season', 'Item Purchased']).size().reset_index(name='count')
item_counts = item_counts.sort_values(['Season', 'count'], ascending=[True, False])
top_items = item_counts.groupby('Season').head(1)
print(top_items)
```

: For fall the most popular item was socks with a frequency of 87.

: For spring running shoes was most popular with a frequency of 106.

: In the Summer shorts was most popular with a frequency 177

: During winter leggings was the most popular item with a frequency of 157.

: It may be benifical for the company to look into promotions and advertising that corresponds to the most popular item purchases for that particular season. For example leggings promotions and advertising during winter or shorts during the summer.

3. What is the effect of promo codes on the dollar amount of purchases?

```python
# Effect of promo codes on purchase amount
promo_effect = shopping_data.groupby('Promo Code Used')['Purchase Amount (USD)'].agg(['count', 'mean', 'median'])
print(promo_effect)
```

: For orders where a promo code wasn't used the average amount per order was $30.17.

: For orders where a promo code was used the average amound per order was $50.02. 

: This tells us the promotions were effective in driving revenue.

#1. Using the purchase amount column which gender spent the most money? This can help determine plans for best market outreach strategies.



```python
gender_spend = shopping_data.groupby('Gender')['Purchase Amount (USD)'].agg(['sum', 'mean','count'])

print(gender_spend)
```

: Males spent the most overall with a sum of $76,032 and 1971 total purchases made

: Females were not far behind in 2nd place with a total of $73620 spent and 1894 total purchases made

: Non-binary's came in 3rd place with a total of $1309 and 35 total purchases made.

: It should be noted that while males spent the most overall and made the most purchases females actually spent the most per purchases with a an average of $38.8 spent per purchase. This is slightly above males who have an average of 38.5 spent per purchase.

: The data points to males and females being virtually tied for the gender most likely to make a purchase and or the highest value purchase. It would be benefical to lean into this and target these demographics via advertising. We can also think about ways to increase Non-binary reputation in our sales data. Maybe a social media collaboration with a popular Non-binary fashion influencer. 

2. Which season did customers spend the most money? This can help determine when should we run promotions? What time of year should we put the most resources into marketing. 

```python
season_pivot = pd.pivot_table(shopping_data, values = 'Purchase Amount (USD)', index = 'Season', aggfunc=['sum', 'mean', 'count'])

print(season_pivot)
```

: The season where customers spent the most money in total was winter with a total of $50681.6

: Fall was 2nd with a total purchase amount of 41,004 but actually had a higher average purchase amount per season $38.8 to winters $38.3

: Winter was carried by having the most total purchases for any season 1321. Fall was 2nd with 1056 total purchases.

: It makes sense to invest in advertising and promotions more heavily during winter and fall as sales are up.

The same findings as above but expressed a a visual.

```python
season_pivot.columns = ['Total Spend', 'Avg Spend', 'Count']
season_pivot = season_pivot.sort_values('Total Spend', ascending=False)
fig, ax = mattyp.subplots(figsize=(8, 5))
ax.bar(season_pivot.index, season_pivot['Total Spend'])
ax.set_title('Total Revenue by Season', fontsize=16)
ax.set_xlabel('Season', fontsize=13)
ax.set_ylabel('Total Spend (USD)', fontsize=13)
mattyp.tight_layout()
mattyp.savefig('revenue_by_season.png', dpi=150)
mattyp.show()
```

This backs up our table from earlier Winter brought in the most revenue followed by fall, summer and spring.

5. Out of our most loyal customers people who made purchases 10 or more times what gender are they most likely to belong to? Again this can help with market outreach strategies? Which group should we target for advertising? What form of advertising would be best for us to invest?

```python
loyal = shopping_data[shopping_data['Previous Purchases'] >= 10]

gender_loyal = loyal['Gender'].value_counts().reset_index()
gender_loyal.columns = ['Gender', 'Count']
print(gender_loyal)
```

: Females represents the most likely group of people to purchase 10 or more items currently but this number is almost 50/50 with males 449 to 444 extremely close.

: This gives strong reason to invest in promotion and advertising evenly between these two groups but earlier numbers suggest it's probably a better idea to invest in one group more heavly depending on the season but evenly overall. Males are more likely to purchase during the fall and females are more likely to purchase during the summer. Advertising can reflect this.
